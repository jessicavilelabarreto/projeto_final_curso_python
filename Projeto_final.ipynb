# import da Base 01 - API Bancos
Tratamentos:
 - Ajuste dos nomes das colunas e linhas.
 - Ajuste de missing

```python
url='https://brasilapi.com.br/api/banks/v1'
response =requests.get(url)
response

if response.status_code==200:
    data_bank=response.json()
else:
    print ("Erro ao acessar a API")

df_banks = pd.DataFrame(data_bank)

#ETL 

# Ajuste os nomes das colunas e linhas.
df_banks.rename(columns={'name':'nome_instituicao_bancaria','fullName':'nome_completo'}, inplace = True)

#Ajuste de missing.

df_banks.fillna({'code':-1})\
        .dropna(subset='nome_completo')
```

# import Base 03 - API CorretorasCVM

```python
import pandas as pd
import requests

url_corretoras='https://brasilapi.com.br/api/cvm/corretoras/v1'
response_corretoras=requests.get(url_corretoras)
response_corretoras

if response_corretoras.status_code==200:
    data_corretoras=response_corretoras.json()
else:
    print ("Erro ao acessar a API")

df_corretoras= pd.DataFrame(data_corretoras)
df_corretoras.head(10)
```

Tratamentos Base 03:
 - Seleção de colunas desejadas da API Corretoras com Status "EM FUNCIONAMENTO NORMAL" Somente
 - Alteração da coluna type para tipo
 - Formatação de campo data_patrimonio_liquido para string

```python
import pandas as pd

# colunas_desejadas
colunas_desejadas = ['nome_social',
                     'cnpj',
                     'codigo_cvm',
                     'type',
                     'municipio',
                     'uf',
                     'data_registro',
                     'status',
                     'data_patrimonio_liquido',
                     'valor_patrimonio_liquido']

# Seleção apenas as corretoras em funcionamento normal
df_Corretoras_funcionamento_normal = df_corretoras[df_corretoras['status'] == 'EM FUNCIONAMENTO NORMAL'][colunas_desejadas]

# Ajustar os nomes das colunas e linhas >> Alteração da coluna 'type' para 'Tipo'
df_Corretoras_funcionamento_normal = df_Corretoras_funcionamento_normal.rename(columns={'type': 'tipo'})

# Formatação do campo data_patrimonio_liquido para string
df_Corretoras_funcionamento_normal['data_patrimonio_liquido'] = pd.to_datetime(df_Corretoras_funcionamento_normal['data_patrimonio_liquido'])
df_Corretoras_funcionamento_normal['data_PL_formatada'] = df_Corretoras_funcionamento_normal['data_patrimonio_liquido'].dt.strftime('%d-%m-%Y')

df_Corretoras_funcionamento_normal.head(10)
```

- Identificando duplicidades:

```python
# Identificar Duplicidades:
duplicatas = df_Corretoras_funcionamento_normal[df_Corretoras_funcionamento_normal.duplicated()]

if duplicatas.empty:
    print("Não há linhas duplicadas no DataFrame df_Corretoras_funcionamento_normal.")
else:
    print("Linhas duplicadas no DataFrame df_Corretoras_funcionamento_normal:")
    print(duplicatas)
```

# import Base 04 - API IBGE - Estados/Regiões - Região Sudeste
Tratamentos:
 - Seleção de dados da Região Sudeste 

```python
import pandas as pd 
regiao= []

# Vamos pegar somente os dados da Região Sudeste == Filtros de linhas e colunas, se necessário. 

for city_sigla in ['SP','MG','RJ','ES']:
    url_região_corretoras =f'https://brasilapi.com.br/api/ibge/uf/v1/'+ str(city_sigla)
    response_região_corretoras =requests.get(url_região_corretoras)
    data_região_corretoras=response_região_corretoras.json()
    regiao.append(data_região_corretoras)

df_regiao_corretoras = pd.DataFrame(regiao)\
                     .drop(columns=['regiao']) 
df_regiao_corretoras['nome regiao'] = data_região_corretoras['regiao']['nome']
df_regiao_corretoras
```
